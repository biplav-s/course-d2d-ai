```python
import pandas as pd
```

```python
# Read some French data
df = pd.read_csv("../common-data/text/french-text.csv", encoding="latin-1")
df
```

```python
# A function to translate from French to English
def translate_fr_en(text, translator):
    result = translator.translate(text, lang_tgt='en')
    # print (result)
    return result
```

```python
# Do translation for the read text
df['en-sentence'] = df.apply(lambda x: translate_fr_en(x['Sentence'], translator), axis=1 )
```
