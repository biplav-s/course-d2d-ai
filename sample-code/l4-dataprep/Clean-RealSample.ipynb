```python
# Making a list of missing value types
missing_values = ["n/a", "na", "--"]
```

```python
df = pd.read_csv("./data/real.csv", na_values = missing_values)
df
```

```python
# See if we just drop all null values
df.dropna()
```

```python
# We can fill missing numeric values with median value
median = df['NUM_BEDROOMS'].median()
df['NUM_BEDROOMS'].fillna(median, inplace=True)
df
```

```python
# We can fill missing numeric values with mean too
mean = df['SQ_FT'].mean()
df['SQ_FT'].fillna(mean, inplace=True)
df
```

```python
# Now see if we just drop all null values
df.dropna()
```
