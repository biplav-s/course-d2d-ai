```python
# Make imports
import requests
import json
```

```python
# Chicago end-point
baseurl = 'http://311api.cityofchicago.org/open311/v2/'
```

```python
# Open311 service / action called
action = 'services' + '.json'  # extension is for result format
```

```python
# A function to make care and get results
def makeRestCall(url):
    
    try:
        res = requests.get(url)
        return res.text

    except Exception as err:
        print(f'Error occurred: {err}') 
        return ""
```

```python
# Print the response result
def printResponse (result):
    j = json.loads(result)
    j_pretty = json.dumps(j, indent=2)
    print (j_pretty)
```

```python
# Make the call to know about Chicago's city services offered
# Doesn't need a API KEY
chicagocall = baseurl + action
result = makeRestCall(chicagocall)
printResponse(result)
```

```python
# Now for SFO
baseurl = 'http://mobile311.sfgov.org/open311/v2/'
action = 'services' + '.json'
```

```python
sfocall = baseurl + action
result = makeRestCall(sfocall)
printResponse(result)
```
