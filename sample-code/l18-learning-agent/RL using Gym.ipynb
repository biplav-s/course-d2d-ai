```python
# Import libs
import gym
```

```python
# Available environments can be printed
from gym import envs

print(envs.registry.all())
```

```python
def run_env(env, episodes, max_timesteps_per_episode):
    
    for i_episode in range(episodes):
        
        observation = env.reset()
        
        for t in range(max_timesteps_per_episode):
            env.render()
            print(observation)
            
            action = env.action_space.sample()
            observation, reward, done, info = env.step(action)
            
            if done:
                print("Episode finished after {} timesteps".format(t+1))
                break
```

```python
default_num_episodes = 1 
default_max_timesteps_per_episode = 5
```

## Run Cartpole

```python
# Details: https://gym.openai.com/envs/CartPole-v1/
```

```python
# Instantiating an environment
env = gym.make("CartPole-v1")
```

```python
# Run for a very short length to illustrate observation at each timestep of episode
run_env(env, default_num_episodes, default_max_timesteps_per_episode)
```

```python
# Run longer: 2 episodes and 100 time steps in each episode
run_env(env, 50, 100)
```

```python
# Close the environment at the end
env.close()
```
